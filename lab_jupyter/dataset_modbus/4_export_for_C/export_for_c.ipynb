### Exporter les données pour le programme C sous nordic

```python
import pandas as pd
```

```python
test_data = pd.read_csv("test_data.csv", low_memory=False)
```

```python
test_data.info()
```

```python
# Filtrer les données
# on veut 10 => '0' et 10 => '1'

normal_rows = test_data[test_data["Normal/Attack"] == 0].head(10)

# Sélectionner 10 lignes avec 'Normal/Attack' = 0
attack_rows = test_data[test_data["Normal/Attack"] == 1].head(10)

# Combiner les deux ensembles
selected_rows = pd.concat([normal_rows, attack_rows])
selected_rows.info()
```

```python
#separation des données
x_test, y_test = selected_rows.drop(columns=['Normal/Attack']), selected_rows['Normal/Attack']
print(x_test.shape)
print(y_test.shape)
```

```python
x_test.info()
```

```python
y_test.info()
```

```python
#transformer en liste

x_test_c = x_test.to_numpy(dtype=float)
y_test_c = y_test.to_numpy(dtype=float)
```

```python
# Générer le code pour C
x_test_c_code = "float x_test_c[20][45] = {\n" + ",\n".join(
    "{" + ", ".join(f"{x:.6f}" for x in row) + "}" for row in x_test_c
) + "\n};"

y_test_c_code = "int y_test_c[20] = {" + ", ".join(map(str, y_test_c)) + "};"

# Afficher le code pour C
print(x_test_c_code)
print(y_test_c_code)
```

```python
# ecriture dans un fichier .c
# Écrire dans un fichier .c
with open("data_test_c.c", "w") as f:
    f.write("// Fichier généré contenant les tableaux x_test_c et y_test_c\n")
    f.write(x_test_c_code + "\n\n")
    f.write(y_test_c_code + "\n")

print("Les tableaux ont été écrits dans le fichier 'output.c'.")
```
